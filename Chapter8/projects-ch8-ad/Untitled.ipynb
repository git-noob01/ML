```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.arange(24).reshape(6,4),index=pd.date_range(start='20170101',periods=6),columns=['A','B','C','D'])
df
```

```python
df['diff'] = df['A'].diff(-1, axis = 0)
df
```

```python
df['diff'] = df['A'].shift(-1) - df['A']
df['label'] = pd.Series(np.zeros(len(df)))

df.loc[df['diff'] > 0,'label'] = 1
df.loc[df['diff'] <= 0,'label'] = 0
df
```
